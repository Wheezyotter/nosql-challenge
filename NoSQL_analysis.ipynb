```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
hygiene_count = establishments.count_documents(hygiene_query)
print(f'There are {hygiene_count} establishements with a hygiene score of 20')

# Display the first document in the results using pprint
pprint(establishments.find_one(hygiene_query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.json_normalize(list(establishments.find(hygiene_query)))
# Display the number of rows in the DataFrame
print(f'There are {len(hygiene_df)} rows')
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
lon_rating_query = {'LocalAuthorityName': {'$regex': 'London'},
                    'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
lon_rating_count = establishments.count_documents(lon_rating_query)
print(f'There are {lon_rating_count} establishements with a RatingValue score >= to 20')
# Display the first document in the results using pprint
pprint(establishments.find_one(lon_rating_query))
```

```python
# Convert the result to a Pandas DataFrame
lon_rating_df = pd.json_normalize(list(establishments.find(lon_rating_query)))

# Display the number of rows in the DataFrame
print(f'There are {len(lon_rating_df)} rows')

# Display the first 10 rows of the DataFrame
lon_rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
query = {'BusinessName': "Penang Flavours"}
fields = {"geocode.latitude": 1, "geocode.longitude": 1}

lat_long = establishments.find(query, fields)
lat_long_df = pd.json_normalize(list(lat_long))
lat_long_df
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = lat_long_df['geocode.latitude'][0]
longitude = lat_long_df['geocode.longitude'][0]

query = {'RatingValue': 5, 
         'geocode': {'$nearSphere': [longitude, latitude],
                                       '$maxDistance': degree_search}}

sort =  [('HygieneScore', -1)]
limit = 5

hygien_rating_query = establishments.find(query).sort(sort).limit(limit)
# Print the results
pprint(list(hygien_rating_query))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
query = {'$match': {'scores.Hygiene': 0}}
group = {'$group': {'_id': '$LocalAuthorityName',
                   'count': { '$sum': 1 }}}
sort = {'$sort': {'count': -1}}

pipeline = [query, group, sort]
auth_hygiene_agg = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of rows in result: ", len(auth_hygiene_agg))

# Print the first 10 results
pprint(auth_hygiene_agg[0:10])
```

```python
# Convert the result to a Pandas DataFrame
auth_hygiene_df = pd.json_normalize(auth_hygiene_agg)
# Display the number of rows in the DataFrame
print(f'There are {len(auth_hygiene_df)} rows')
# Display the first 10 rows of the DataFrame
auth_hygiene_df.head(10)
```
